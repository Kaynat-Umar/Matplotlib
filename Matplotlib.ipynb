```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.arange(-10, 11)
b=a**2
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, figsize= (8,8))

ax1.set_title("subplot 1")
ax2.set_title("subplot 2")
ax3.set_title("subplot 3")
ax4.set_title("subplot 4")

ax1.plot(a, b, 'go')
ax2.plot([1, 2, 3, 4], [1, 2, 3, 4])
ax3.plot([0, 0, 0], [-10, 0, 100])
ax4.plot(a,b)
ax4.plot([0, 0, 0], [-10, 0, 100])
```

```python
plt.figure(figsize=(5,5))
plt.title('My Second Plot')

plt.subplot(1, 2, 1)
plt.plot(a, b)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['A^2', 'Vertical Line'])
plt.xlabel('A')
plt.ylabel('A Squared')

plt.subplot(1, 2, 2)
plt.plot(a, -1 * b)
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-A^2', 'Horizontal Line'])
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, figsize= (15,15))

ax1.set_title("subplot 1")
ax2.set_title("subplot 2")
ax3.set_title("subplot 3")
ax4.set_title("subplot 4")
plt.figure()
ax1.scatter(x,y, s=area, c= colors, alpha=0.8, cmap= "gist_rainbow")
ax2.scatter(x,-y, s=area, c= colors, alpha=0.8, cmap= "gnuplot")
ax3.scatter(-x,-y, s=area, c= colors, alpha=0.8, cmap= "cividis")
ax4.scatter(-x,y, s=area, c= colors, alpha=1, cmap= "CMRmap")
```

```python
x=np.random.rand(50)
y=np.random.rand(50)

colors= np.random.rand(50)
area = np.pi * (20 * np.random.rand(50)) **2

plt.figure(figsize=(10, 6))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))

ax1.scatter(x,y, s=area, c=colors, alpha = 0.6, cmap="gist_rainbow")
ax2.plot(x, c='blue', linestyle=':')
ax3.scatter(y,x, s=area, c=colors, alpha = 0.8, cmap="turbo")
ax4.plot(np.random.randn(25), c='purple')
ax5.plot(np.random.randn(30), c='green', marker='o', linewidth=3.0)
```

```python
#concatenating two separate arrays, one with 10 random values and one with 10,15,-10 and -15, which will be the outliers

values = np.concatenate([np.random.randn(10), np.array([10, 15, -10, -15])])

#first using a histogram to visualize outliers
plt.figure(figsize=(8, 4))

plt.hist(values)
```

```python
#now using a boxplot to visualize outliers

plt.figure(figsize=(8,4))

plt.boxplot(values)
```
